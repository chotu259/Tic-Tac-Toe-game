```python
def currentboard(board):
  print("current board")
  for i in range(0,9):
    if(i>0 and i%3==0):
      print('\n')
    if(board[i]==0):
      print("_",end=' ')
    if(board[i]==-1):
      print("X",end=' ')
    if(board[i]==1):
      print("O",end=' ')
  print("\n\n")
```

```python
def analyzeboard(board):
  cb=[[0,1,2],[3,4,5],[6,7,8],[0,3,6],[1,4,7],[2,5,8],[2,4,6],[0,4,8]]
  for i in range(0,8):
    if(board[cb[i][0]]!=0 and board[cb[i][0]]==board[cb[i][1]] and board[cb[i][0]]==board[cb[i][2]]):
      return board[cb[i][0]]
  return 0;
```

```python
def user1turn(board):
  pos=int(input("Enter the value of X position between [1,2,3...,9] : "))
  if(board[pos-1]!=0):
    print("wrong move")
    exit(0);
  board[pos-1]=-1;


def user2turn(board):
  pos=int(input("Enter the value of O position between [1,2,3...,9]: "))
  if(board[pos-1]!=0):
    print("wrong move")
    exit(0);
  board[pos-1]=1;
```

```python
def minmax(board,player):
  x=analyzeboard(board)
  if(x!=0):
    return (x*player)
  pos=-1;
  value=-2;
  for i in range(0,9):
    if(board[i]==0):
      board[i]=player
      score=-minmax(board,-1*player)
      board[i]=0
      if(score>value):
        value=score;
        pos=i;
  if(pos==-1): return 0;
  return value

def computerturn(board):
  pos=-1;
  value=-2;
  for i in range(0,9):
    if(board[i]==0):
      board[i]=1
      score=-minmax(board,-1)
      board[i]=0
      if(score>value):
        value=score;
        pos=i;
  board[pos]=1
```
